#**Import Pre-Trained Model**

```python
#Load the EfficientNetB0 model
effnet = efficientnet.EfficientNetB0(weights='imagenet', 
                 include_top=False, 
                 input_shape=(256, 256, 3))
effnet.trainable = False
```

#**Create Sequential Model**

```python
# Create the model
model = models.Sequential()
 
# Add the efficientnetB0 convolutional base model
model.add(effnet)

# Add new layers
model.add(layers.Flatten())
#model.add(layers.GlobalAveragePooling2D())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))

model.add(layers.Dense(1, activation='sigmoid'))
 
# Show a summary of the model. Check the number of trainable parameters
model.summary()
```


#**Data Augmentation and Data Generation**


**Create Function for Hue Augmentation**

```python
def augment_hue(image):
    # Convert image from BGR to HSV
    image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Generate random hue adjustment
    hue_adjustment = np.random.uniform(-5, 10)

    # Add hue adjustment to image
    image[:, :, 0] = image[:, :, 0] + hue_adjustment

    # Convert image back to BGR
    image = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)

    return image

#Function to combine hue augmentation and EfficientNetB0 preprocess_input
def preprocess(image):
    # Apply hue adjustment
    image = augment_hue(image)
    # Apply preprocessing
    image = preprocess_input(image)
    return image
```

```python
my_model = models.Sequential()

my_model.add(layers.Conv2D(64, kernel_size=(5, 5), activation='relu', padding="same", input_shape=(256, 256, 3)))
my_model.add(layers.Conv2D(64, kernel_size=(3, 3), activation='relu', padding="same"))

my_model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

my_model.add(layers.Conv2D(128, kernel_size=(5, 5), activation='relu', padding="same"))
my_model.add(layers.Conv2D(128, kernel_size=(3, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(128, kernel_size=(1, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(128, kernel_size=(3, 1), activation='relu', padding="same"))
my_model.add(layers.Conv2D(128, kernel_size=(3, 3), activation='relu', padding="same"))

my_model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

my_model.add(layers.Conv2D(256, kernel_size=(3, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(256, kernel_size=(1, 1), activation='relu', padding="same"))
my_model.add(layers.Conv2D(256, kernel_size=(3, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(256, kernel_size=(1, 1), activation='relu', padding="same"))
my_model.add(layers.Conv2D(256, kernel_size=(3, 3), activation='relu', padding="same"))

my_model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

my_model.add(layers.Conv2D(512, kernel_size=(3, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(512, kernel_size=(3, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(512, kernel_size=(3, 3), activation='relu', padding="same"))

my_model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

my_model.add(layers.Conv2D(512, kernel_size=(3, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(512, kernel_size=(3, 3), activation='relu', padding="same"))
my_model.add(layers.Conv2D(512, kernel_size=(3, 3), activation='relu', padding="same"))

my_model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

my_model.add(layers.Flatten())

my_model.add(layers.Dense(256, activation='relu'))

# my_model.add(layers.Dropout(0.5))

my_model.add(layers.Dense(512, activation='relu'))

# my_model.add(layers.Dropout(0.5))

my_model.add(layers.Dense(1, activation='sigmoid'))

my_model.summary()
```
